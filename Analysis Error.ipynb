```python
def normalize_address(addr):
    if not isinstance(addr, str):
        return str(addr)
    addr = ' '.join(addr.split(','))
    addr = addr.lower()
    addr = ' '.join(addr.split())
    return addr.strip()
```

```python
df['pd_home_norm'] = df['pd_home'].apply(normalize_address)
```

```python
df['gt_home_norm'] = df['gt_home'].apply(normalize_address)
```

```python
def correct_address(address):
    correct_address = address_correction.address_correction(address)
#     if correct_address[1] != 0:
#         print(address)
#         print(correct_address)
#         print()
    return correct_address
```

```python
df['correct_pd_home'] = df['pd_home_norm'].apply(correct_address)
```
